**Question 4** Another important part of EDA is to make data usable. For example, the `season` column contains strings of the form '2023' or '2023/24', etc.; these as integers would open them up for further analysis. Therefore, change the values in this column to integers (all seasons of the form 2023/24 should be considered as 2024 and so on).

```python
def transform_season(x):
        # If x is a string and contains '/', split it and take the second part
        if isinstance(x, str) and '/' in x:
            return int("20" + x.split('/')[1])
        # If x is already an integer or a string without '/', assume it's the year
        else:
            return int(x)
# Apply the function to the 'season' column
cricketData['season'] = cricketData['season'].apply(transform_season)

cricketData
```

**Question 8** Next let's move on to investigate the relationship between winning the toss and winning the match, just like before create a dictionary that has the team name as key, and the probability of winning the match given the team have won the toss as value (make sure to sort it in alphabetic order of team name). Then plot your results; justify the choice of plot. (For this question, you must decide whether all teams should be included or not)

```python
# Calculate wins after winning the toss
toss_wins = cricketData['TossWinner'].value_counts()  # Total toss wins per team
wins_after_toss = cricketData[cricketData['TossWinner'] == cricketData['Winner']]['Winner'].value_counts()  # Wins after toss

# Calculate probabilities
# P(Win | Toss) = (Wins after Toss) / (Total Toss Wins)

probabilities = {team: (wins_after_toss.get(team, 0) / toss_wins[team]) for team in toss_wins.index}

# Sort dictionary by team name
sorted_probabilities = dict(sorted(probabilities.items()))

# Plot the results
plt.figure(figsize=(10, 6))
plt.bar(sorted_probabilities.keys(), sorted_probabilities.values(), color='green', edgecolor='black')

# Add labels and title
plt.title('Probability of Winning the Match Given Toss Won (by Team)', fontsize=14)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Probability of Winning', fontsize=12)

# Display the plot
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Bar Plot** is suitable because:

It compares probabilities across categories (teams).
Probabilities are numerical values, and bar height makes differences visually distinct.

Plot this information and now comment on who might be the top team (store the name in `actualBestTeam`) and whether this is a better comparison than what we did before.

**Question 16** For all seasons, find the player that has been the most frequent Man of the match for Pakistan. There might be people with the same number of wins, in that case give sequential preference. Store in a dictionary, where key is season and value is the name of the player.

```python
#PakManOfMatchSeason

# Step 1: Filter matches won by Pakistan
pak_wins = cricketData[cricketData['Winner'] == 'Pakistan']

# Step 2: Group by season and player to count Man of the Match awards
player_counts = pak_wins.groupby(['Season', 'PlayerOfMatch']).size().reset_index(name='count')

# Step 3: Find the player with the most awards per season
PakManOfMatchSeason = {}
for season, group in player_counts.groupby('Season'):
    # Find the player(s) with the maximum count
    top_player = group[group['count'] == group['count'].max()].iloc[0]  # Sequential preference
    PakManOfMatchSeason[season] = top_player['PlayerOfMatch']

# Display results
print("Most Frequent Man of the Match for Pakistan by Season:", PakManOfMatchSeason)
```

**Question 17** Lets now analyze more recent performance of the teams. For the last 3 seasons, plot the distribution of `WinByRuns` for the top 10 teams with most wins. For each time, overlay the distribution of each season on the same plot. Comment on what you observe.

```python
# Step 1: Filter for recent 3 seasons
recent_seasons = cricketData['Season'].unique()[-3:]  # Last 3 unique seasons
recent_data = cricketData[cricketData['Season'].isin(recent_seasons)]

# Step 2: Calculate total wins for each team and find top 10
top_teams = (
    recent_data['Winner']
    .value_counts()
    .head(10)
    .index
)

# Step 3: Filter data for top 10 teams
top_teams_data = recent_data[recent_data['Winner'].isin(top_teams)]

# Step 4: Plot histograms
plt.figure(figsize=(14, 10))

for team in top_teams:
    team_data = top_teams_data[top_teams_data['Winner'] == team]
    for season in recent_seasons:
        season_data = team_data[team_data['Season'] == season]
        plt.hist(season_data['WinByRuns'], bins=10, alpha=0.5, label=f'{team} ({season})', histtype='stepfilled')

# Step 5: Customize plot
plt.xlabel('Win Margin (Runs)')
plt.ylabel('Frequency')
plt.title('Win Margin Distribution (Runs) for Top 10 Teams Over Recent 3 Seasons')
plt.legend(title='Team and Season', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Question 18** Similarly, for all seasons, find the team that Pakistan has defeated the most frequently. Store your results in a dictionary called `mostDefeatedTeam`, where key is the season and the team name is value.

```python
#mostDefeatedTeam

# Step 1: Filter matches won by Pakistan
pakistan_wins = cricketData[cricketData['Winner'] == 'Pakistan']

# Step 2: Group by season and team2 (opponent)
win_counts = (
    pakistan_wins.groupby(['Season', 'Team2'])
    .size()
    .reset_index(name='win_count')
)

# Step 3: Identify the most defeated team for each season
mostDefeatedTeam = (
    win_counts.loc[win_counts.groupby('Season')['win_count'].idxmax()]
    .set_index('Season')['Team2']
    .to_dict()
)

# Display the result
print("Most Defeated Teams by Pakistan (per Season):", mostDefeatedTeam)
```
